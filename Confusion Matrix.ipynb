```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

array = [[411,25,16,13], 
        [25,407,18,20], 
        [17,21,409,26], 
        [15,16,22,415]]

np.set_printoptions(precision=4)

df_cm = pd.DataFrame(array, index = [i for i in ["EOSINOPHIL", "LYMPHOCYTE", "MONOCYTE", "NEUTROPHIL"]],
                  columns = [i for i in ["EOSINOPHIL", "LYMPHOCYTE", "MONOCYTE", "NEUTROPHIL"]])

plt.figure(figsize = (10,8))
sn.heatmap(df_cm, annot=True, cmap="Blues", fmt=".0f")
plt.title('Confusion Matrix for CNN Model with Maximum Training Images')
plt.xlabel('Predicted Label')
plt.xticks(rotation=45)
plt.ylabel('True Label')
```
